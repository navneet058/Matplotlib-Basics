# This section gives the basic overview of matplotlib library

```python
#Import the matplotlib library

import numpy as np
import matplotlib.pyplot as plt

from matplotlib.ticker import NullFormatter 
```

```python
#Make the simple linear plot with list value

"""
You may be wondering why the x-axis ranges from 0-4 and the y-axis from 1-5. 
If you provide a single list or array to the plot() command, matplotlib assumes it is a sequence of y values, and automatically generates the x values for you. 
Since python ranges start with 0, the default x vector has the same length as y but starts with 0. Hence the x data are [0,1,2,3,4].

"""



plt.plot([1,2,3,4,5])
plt.xlabel('Distance Travelled')
plt.ylabel('Time elasped')
plt.show()
```

```python
"""
The first argument represents the x axis and second argument represents the y-axis

"""
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
```
